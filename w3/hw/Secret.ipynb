```python
pokemon = pd.read_csv("pokemon.csv")
pokemonName = pokemon["Name"]
pokemonName.to_csv('pokemon_name.csv', index=False)
```

```python
import csv
import requests
import time

base_url = "https://api-kmuttme.kmutt.ac.th/label/NWartyin3055LN455"
input_csv = "pokemon_name.csv"
output_csv = "found_results.csv"  # ไฟล์ที่จะบันทึกผลลัพธ์ที่สำเร็จ

def brute_force():
    try:
        # เปิดไฟล์ Input เพื่ออ่านข้อมูล
        with open(input_csv, mode='r', encoding='utf-8') as file:
            reader = csv.reader(file)
            
            for row in reader:
                if not row or not row[0].strip():
                    continue
                
                # จัดการ String ให้ติดกันและเป็นตัวเล็ก
                clean_payload = row[0].replace(" ", "").lower()
                target_url = f"{base_url}{clean_payload}"
                
                print(f"กำลังทดสอบ: {target_url}")
                
                try:
                    response = requests.get(target_url, timeout=5)
                    
                    if response.status_code == 200:
                        print(f"✅ [สำเร็จ] พบข้อมูล: {clean_payload}")
                        
                        # --- ส่วนที่เพิ่ม: บันทึกลง CSV ทันทีที่เจอ ---
                        with open(output_csv, mode='a', encoding='utf-8', newline='') as outfile:
                            writer = csv.writer(outfile)
                            # บันทึกค่าที่ใช้ได้ และ Response ที่ได้จาก Server
                            writer.writerow([clean_payload, response.text])
                        # ---------------------------------------
                        
                    else:
                        print(f"❌ [ล้มเหลว] Status: {response.status_code}")
                
                except requests.exceptions.RequestException as e:
                    print(f"⚠️ เกิดข้อผิดพลาดในการเชื่อมต่อ: {e}")
                
                # หน่วงเวลา 0.3 วินาที
                time.sleep(0.5)

    except FileNotFoundError:
        print(f"ไม่พบไฟล์ {input_csv}")

if __name__ == "__main__":
    # สร้างหัวตาราง (Header) ให้ไฟล์ output ก่อนเริ่มทำงาน (ถ้ายังไม่มีไฟล์)
    try:
        with open(output_csv, mode='x', encoding='utf-8', newline='') as f:
            writer = csv.writer(f)
            writer.writerow(['Payload', 'Response_Data'])
    except FileExistsError:
        pass # ถ้ามีไฟล์อยู่แล้วไม่ต้องสร้างหัวใหม่

    brute_force()
```
